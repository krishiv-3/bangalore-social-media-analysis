```python
with open('finaldata.txt' , encoding='utf-8') as f:
    data = f.read()
```

```python
# data
```

```python
chunks = data.split('\n\n')
chunks = [c for c in chunks if len(c)>3]
```

```python
# chunks[0]
```

```python
def parseChunks(chunk) :
   
         chunk = chunk.strip()
         sep_chunk = chunk.split('\n')
         username = sep_chunk[0]
         no_of_posts = int(sep_chunk[1].split(" post")[0].replace("," ,""))
    
         no_of_followers = float(sep_chunk[2].split(" followers")[0].replace("," ,"").replace("K" , "").replace("M" , ""))
         if ("K" in sep_chunk[2]):
             no_of_followers = int(no_of_followers * 1000)
         elif ("M" in sep_chunk[2]):
             no_of_followers = int(no_of_followers * 1000000)
         else  :
             no_of_followers = int(no_of_followers)
        
         no_of_following = float(sep_chunk[3].split(" following")[0].replace("," ,"").replace("K" , "").replace("M" , ""))
         if ("K" in sep_chunk[3]):
             no_of_following = int(no_of_following * 1000)
         elif ("M" in sep_chunk[3]) :
             no_of_following = int(no_of_following * 1000000)
         else :
             no_of_following = int(no_of_following)
             
         name = sep_chunk[4]
    
         if(len(sep_chunk)>5):
             category = sep_chunk[5]
             bio = sep_chunk[6:]
         else :
             category ="unknown"
             bio = ""
         # print(username ,no_of_posts , no_of_followers , no_of_following , name ,category , bio , sep='\n')
         return { "username" : username ,"no_of_posts" : no_of_posts ,"no_of_followers" : no_of_followers , "no_of_following" : no_of_following ,"name" : name ,"category" : category ,"bio" : bio  }

# print(parseChunks(chunks[1]))
```

```python
all_chunks = []
for chunk in chunks :
    parsed_chunk = parseChunks(chunk)
    all_chunks.append(parsed_chunk)
# print(all_chunks)
```

```python
import json
```

```python
s = json.dumps(all_chunks , indent = 4)
with open("data.json" , "w") as f:
    f.write(s)
# print(s)
```

```python
# Who has the maximum posts?
max_POSTS = 0
for chunk in all_chunks :
    if chunk["no_of_posts"] > max_POSTS:
        max_POSTS = chunk["no_of_posts"]
        personWithMaxPosts = chunk
        # print(chunk["no_of_posts"])
print(personWithMaxPosts['username'])
```

```python
# Who has the maximum FOLLOWERS?

maxFollowers = 0
for chunk in all_chunks :
    if chunk["no_of_followers"] > maxFollowers:
        maxFollowers = chunk["no_of_followers"]
        personWithMaxFollowers = chunk

print(personWithMaxFollowers['username'])
```

```python
maxFollowing = 0
for chunk in all_chunks :
    if chunk["no_of_following"] > maxFollowing:
        maxFollowing = chunk["no_of_following"]
        personWithMaxFollowing = chunk

print(personWithMaxFollowing['username'])
```

```python
categories = []
for chunk in all_chunks:
    if chunk["category"] is None :
        chunk["category"] = "unknown"
    if chunk["category"] not in categories:
        categories.append(chunk["category"])
    # print(chunk["category"])
print(f"there are {len(categories)} categories")
```
